# No Memory

```python
from langchain.agents import create_agent

agent = create_agent("gpt-5-nano")
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content = "Hello My Name is FilledVaccum, and my favorite color is Black.")

response = agent.invoke(
    { "messages" : [question] }
)
```

```python
from pprint import pprint
pprint(response)
```

```python
question = HumanMessage(content="What's my favorite color?")

response = agent.invoke(
    { "messages" : [question] }
)

pprint(response)
```

## Memory

```python
from langgraph.checkpoint.memory import InMemorySaver

agent = create_agent(
    "gpt-5-nano",
    checkpointer=InMemorySaver(),
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content = "Hello My Name is FilledVaccum and my fav color is Black?")

config = { "configurable" : {"thread_id" : "1"} }

response = agent.invoke(
    {"messages": [question] },
    config,
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="What is my favorite color?")

response = agent.invoke(
    { "messages" : [question] },
    config,
)

pprint(response)
```
