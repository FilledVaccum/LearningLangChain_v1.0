```python
import subprocess
import threading

def run_sqlite_web():
    subprocess.run(['sqlite_web', 'chatbot.db', '--host', '0.0.0.0', '--port', '8080'])

thread = threading.Thread(target=run_sqlite_web, daemon=True)
thread.start()

print("SQLite viewer running at: https://<your-sagemaker-domain>.studio.<region>.sagemaker.aws/jupyter/default/proxy/8080/")
```
